```python
from datasets import load_dataset

ds = load_dataset("swaption2009/20k-en-zh-translation-pinyin-hsk")
```

```python
print(ds)
```

```python
import pandas as pd

texts = ds['train']['text']

# Each record is 5 consecutive rows: english, hsk, mandarin, pinyin, separator
records = []
for i in range(0, len(texts), 5):
    english = texts[i].replace('english: ', '', 1)
    hsk_level = int(texts[i + 1].replace('hsk: ', '', 1))
    mandarin = texts[i + 2].replace('mandarin: ', '', 1)
    pinyin = texts[i + 3].replace('pinyin: ', '', 1)
    records.append({
        'english': english,
        'hsk_level': hsk_level,
        'chinese': mandarin,
        'pinyin': pinyin,
    })

df = pd.DataFrame(records)

# Filter for HSK levels 2 and 3
df = df[df['hsk_level'].isin([2, 3])].reset_index(drop=True)

print(f"Total records: {len(df)}")
df.head(10)
```

```python
df.to_csv("chinese_data.csv", index=False)
```
