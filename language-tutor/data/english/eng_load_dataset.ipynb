```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("UniversalCEFR/readme_en")
```

```python
print(ds)
```

```python
# Convert a specific split to Pandas
df = ds['train'].to_pandas()

# Filter for A1 and A2 CEFR levels only
df = df[df['cefr_level'].isin(['A1', 'A2'])]

df.head()
```

```python
import pandas as pd

df.to_csv("english_data.csv", index=False)
```
